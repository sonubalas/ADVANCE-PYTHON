Flirting ,Sorting and Mofifying Data

```python
data={"Name":['Anjali',"Bubbles","Rahul",'Priya','Newton'],
     "Age":[23,26,20,27,35],
     "salary":[50000,60000,55000,65000,70000]}
df=pd.DataFrame(data)   # keeping labels
print(df)
```

Filetring

```python
filtered_data=df[df['Age'] > 25]      ## selecting rows where age> 25
print(filtered_data)
```

 using  & (AND) AND / (OR)

```python
filtered_data=df[(df['Age'] > 25)  &  (df['salary'] < 45000)]  # using the AND operator   
print(filtered_data)

filtered_data=df[(df['Age'] > 25)  |  (df['Age'] < 22)]  # using the AND operator   
print(filtered_data)
```

```python
# shorting data


sorted_df=df.sort_values('Age')   # sorting according to AGe col
print(sorted_df)

# descending order of sort using ascending = False parameter

sorted_df= df.sort_values('Name' , ascending=False)  # sorting according to Name col
print(sorted_df)

sorted_df= df.sort_values(['Age' , 'Name'] , ascending= [True ,False])  # sorting according to Name col
print(sorted_df)
```

Handle Missin Fg Values Once you're idenfified columns with missing values you can 
choose hwo to handle them . These are several apporaches:
    Fill missing vlaues for numeric columns you can fill missing values with the mean or median.
    For categorical columns you can fill with the mode or a placeholder (ex. Unknown)
    Drop missing values if a  column or row has too many missing values it may be better to drop it.

```python
# fill missing values in rating col with unknow
dataset['cough']  = dataset['cough'].fillna('Unknown')
print(dataset.isnull().sum())
```

```python
# dropping rows where direction or cast are missing

dataset =dataset.dropna(subset = ['city' , 'has_covid'])
print(dataset.isnull().sum())
```

```python
# ensure cols have correct data type

print(dataset.dtypes)
```

```python
# removing rows where data_added is missing/

dataset=dataset.dropna(subset='["date_added"])
print(dataset.isnull().sum())
```

```python
# cheching col name
print(dataset['gender'].head())
```

```python
dataset['gender'] = dataset['gender'].fillna('others')    # filling with
# others in missing values
print(dataset.isnull().sum())
```
