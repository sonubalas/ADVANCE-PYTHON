```python
import matplotlib.pyplot as plt
```

```python
x=("python" ,"sequal" ,"c" ,"c++" ,"java")
y=(90,75,45,55,80)
plt.ylabel("Number in percent")
plt.title("Language in percentage",fontsize=20)
c=["r","m","g","b","y"]
plt.bar(x,y,color=c,width=0.4,align='edge',edgecolor='b',linewidth=1,alpha=.8,label="language")
plt.legend()
plt.show()
```

```python
x=("python" ,"sequal" ,"c" ,"c++" ,"java")
y=(90,75,45,55,80)
z=[10,20,30,40,50]
plt.ylabel("Number in percent")

plt.title("Language in percentage",fontsize=15)
c=["r","m","g","b","y"]
plt.bar(x,y,color=c,align='center',linewidth=2,label="language",width=0.4)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

x = ['Kotputli', 'Jaipur', 'Alwar', 'Ajmer', 'Sikar']
y = [55, 90, 78, 60, 82]  
colors = ['r', 'g', 'y', 'b', 'g']
plt.bar(x, y, color=colors,width=0.5,label="city-values")
plt.xlabel("Cities",fontsize=16)
plt.ylabel("Values",fontsize=16)
plt.title("City-wise Values", fontsize=15)
plt.legend()
plt.show()
```

```python
x=["chemistry","physics","biology","English"]
y=[74,75,86,60]
plt.xlabel("Subject")
plt.ylabel("Marks")
plt.title("Marks Sheet")
c=["r","y","g","b"]
plt.barh(x,y,color=c,label="Subject_Wise_Marks")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
```

```python
day=[1,2,3,4,5,6,7]
No=[2,7,1,11,9,9,15]
plt.xlabel("day",fontsize=16)
plt.ylabel("Number",fontsize=16)
plt.title("Number wise day",fontsize=16)
plt.scatter(day, No)
plt.show()
```

```python
day=[1,2,3,4,5,6,7]                                                                # marker=change the desine of graph
No=[2,4,1,14,7,9,11]                                                               #edgecolor= give the different type of colours
plt.xlabel("day",fontsize=16)                                                     # aplha graph transparancy can be change(range = 0-1)
                                                                                 #
plt.ylabel("Number",fontsize=16)
plt.title("Number wise day",fontsize=16)
colors=["r","g","y","b","g","m","b"]
size=[300,150,200,400,230,150,300]
plt.scatter(day, No,c=colors,s=size,alpha=0.6,marker='*',edgecolor="r")
plt.show()
```

```python
import matplotlib.pyplot as plt
```

```python
x=[10,30,78,25,55]
y=["python","java","c","c++","sql"]

c=["r","g","y","b","m"]
ex=[0.2,0.0,0.1,0.0,0.0]

plt.pie(x,labels=y,colors=c,explode=ex,autopct="%0.2f%%",shadow=True,radius=1.2,labeldistance=1.1,
       textprops={"fontsize" : 13})

plt.title("Languages In Percentage")
plt.legend(loc=1)
plt.show()
```

```python
plt.pie([2])
plt.show()
```

```python
x=[10,30,78,25,55]
y=["python","java","c","c++","sql"]
y1=["jaipur","kotputli","ajmer","siker","kota"]
z=[12,34,56,78,45]
c=["r","g","y","b","m"]

plt.pie(x,labels=y,radius=1.5,autopct="%0.2f%%")
plt.pie(z,radius=0.7,labels=y1,colors=c,autopct="%0.2f%%")
plt.title("Languages In Percentage")
plt.show()
```
