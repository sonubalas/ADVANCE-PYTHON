```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd



# by use of seaborn

var=[1,2,3,4,5,6,7]
var1=[1,2,3,4,5,2,6]

x1=pd.DataFrame ({"var":var , "var1":var1})
sns.lineplot(x="var" , y="var1" ,data=x1)

plt.show()
```

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x=sns.load_dataset("penguins")
x
```

```python
sns.displot(x["flipper_length_mm"],bins=[170,180,190,200,210,220,230,240],kde=[True],color="g")

plt.show()
```
